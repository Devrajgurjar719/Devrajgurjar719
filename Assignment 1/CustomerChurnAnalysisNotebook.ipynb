```python
#Merging all the datasets to get the final dataset
merge_result1 = pd.merge(cust_demographic, cust_Investment,  how = 'outer', on = ['RowNumber','CustomerId'])
merge_result2 = pd.merge(merge_result1, cust_attrittion_status,  how = 'outer', on = ['RowNumber','CustomerId'])
merge_result3 = pd.merge(merge_result2, cust_portfolio,  how = 'outer', on = ['RowNumber','CustomerId'])
print(merge_result1)
print(merge_result2)
print(merge_result3)
```

```python
#Desired final column positions
col_name_list= ['RowNumber',
 'CustomerId',
 'Surname',
 'CreditScore',
 'Geography',
 'Gender',
 'Age',
 'Tenure',
 'Balance',
 'NumOfProducts',
 'HasChckng',
 'IsActiveMember',
 'EstimatedSalary',
 'Exited']
```

```python
#Altering the final dataset column positions as per the requirement of the project solution
final_result = merge_result3[col_name_list]
final_result
```

```python
final_result.dtypes
```

```python
#Changing the datatypes as asked in the project submission guidlines
final_result['CreditScore'] = final_result['CreditScore'].astype('int64')
final_result['CustomerId'] = final_result['CustomerId'].astype('int64')
final_result['Age'] = final_result['Age'].astype('int')
final_result['Tenure'] = final_result['Tenure'].astype('int')
final_result['NumOfProducts'] = final_result['NumOfProducts'].astype('int')
final_result['HasChckng'] = final_result['HasChckng'].astype('int')
final_result['IsActiveMember'] = final_result['IsActiveMember'].astype('int')
final_result['Exited'] = final_result['Exited'].astype('int')
```

```python
#Filtering the dataset based on credit score as suggested in the project submission guidlines.
final_result = final_result[(final_result['CreditScore']>=350)  & (final_result['CreditScore']<=850)]
```

```python
final_result
```

```python
#Correcting the worng age values in the dataset i.e making sure that no age should be more than 105.
#there was an age with value 192 so replacing that with max value 105
final_result['Age'] = np.where(final_result['Age'] >= 105, final_result["Age"].mean(), final_result['Age']) 
```

```python
#Count of Exited/Non Exited Customers based on Geography
CR_not_exited=grpby_Geography['Central'][0]
ER_not_exited=grpby_Geography['East'][0]
WR_not_exited=grpby_Geography['West'][0]

geographically_non_exited_customers= [WR_not_exited,CR_not_exited,ER_not_exited]
geographically_exited_customers=[WRE,CRE,ERE]
print('Count of geographically Non Exited Customers:- ',geographically_non_exited_customers)
print('Count of geographically Exited Customers:- ',geographically_exited_customers)
```

```python
#Bar Graph of Exited/NonExited Customers based on Geography
geographic_regions = final_result["Geography"].unique()
X_axis = np.arange(len(geographic_regions))
plt.bar(X_axis-0.2, geographically_non_exited_customers, 0.4, label = '0')
plt.bar(X_axis+0.2, geographically_exited_customers, 0.4, label = '1')
plt.xticks(X_axis, geographic_regions)
plt.title('Exited Customers vs Non Exited Customers In Each Geographical Region')
plt.ylabel('Customer Count')
plt.xlabel('Gergraphical Region')
plt.legend(title = "Exited")
```
